```python
#criando nosso Dataframe
df = pd.read_excel("AdventureWorks.xlsx")
df.head(3)
```

```python
#quantidade de linhas e colunas(.shape)
df.shape 
```

```python
#verificando os tipos de dados(.type)
df.dtypes
```

```python
#Calula primeiro o custo total 
df["Custo total"] = df["Custo Unitrio"].mul(df["Quantidade"])
df.head(1)
```

```python
#qual a receita total 
df["Valor Venda"].sum()
```

```python
#Custo total 
df["Custo total"].sum()
```

```python
#Temos a receita e custo e o total, podemos achar o lucro total
df["Lucro total"] = df["Valor Venda"] - df["Custo total"]
df.head(1) 
```

```python
#total lucro 
round(df["Lucro total"].sum(), 2)
```

```python
#tempo para enviar um produto
df["Tempo de envio"] = df["Data Envio"] - df["Data Venda"]
df.head(1)
```

```python
#Extraindo apenas os dias
df["Tempo de envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
df.head(1)
```

```python
#Mdi do tempo de envio por marcar
df["Tempo de envio"].mean()
```

```python
#Mdi do tempo de envio por marcar(aqui  precio combinar as colunas por meio do .groupby)
df.groupby("Marca")["Tempo de envio"].mean()
```

```python
#valores nulos
df.isnull().sum()
```

```python
df.groupby("Marca")["Tempo de envio"].mean()
```
